```python
def sumar(a,b):
   """Devuelve la suma de dos números """
   resultado = a + b
   return resultado

suma = sumar(5, 3)
print(f"la suma es: {suma}")  # salida: la suma
```

```python
def es_par(numero):
  """Devuelve True si el número es par, false si es impar. """
  return numero % 2 == 0

resultado = es_par(4)
print(resultado)  # salida: True
```

```python
def potencia(base, exponente=2):
  return base ** exponente

print(potencia(5))    # usa el exponente por defecto (2), Salida: 25
print(potencia(5, 3))   # usa el exponente proporcionando, Salida: 125
```

```python
def imprimir_info(**datos):
  for clave, valor in datos.items():
    print(f"{clave}: {valor}")
  print(datos)

imprimir_info(nombre="Laura", edad=28, ciudad="Madrid")

```

```python
doblar = lambda x: x * 2
print(doblar(5))  # salida: 10

#uso con funciones de orden superior como map(), filter
numeros=[1,2,3,4,5]
doblados=list(map(lambda x: x * 2, numeros))
print(doblados)   # salida: [2,4,6,8,10]
```

```python
def factorial(n):
  if n == 0 or n ==1:
    return 1
  else:
    return n * factorial(n -1)

print(factorial(5))
```

```python
import math
def area_circulo(radio):
  """
  Calcula el área de un circulo dado su radio.

  Parámetros:
  radio (float): el radio del círculo.

  Retorna:
  float: el área del círculo.
  """
  return math.pi * radio ** 2

print(area_circulo(5))
```

```python
def validar_edad(edad):
  return edad >= 18

def validar_pais(pais):
  return pais.lower() in ["argentina", "chile", "colombia"]

nombre=input("ingresa tu nombre: ")
edad=int(input("ingresa tu edad: "))
pais=input("ingresa tu pais: ").lower()
if validar_edad(edad) and validar_pais(pais):
  print(f"Hola {nombre}, Ud puede a acceder a un beneficio por tener {edad} años y vivir en {pais}.")
else:
  print(f"Hola {nombre}, Ud no puede a acceder a un beneficio por no tener 18 años o porque no vive en Chile, Argentina o Colombia.")

```
